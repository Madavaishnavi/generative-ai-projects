```python
import os
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
import pdfplumber

# Set your OpenAI API key
os.environ["OPENAI_API_KEY"] = "OpenAI API key"

# Function to extract text from a single PDF
def pdftotext(pdf_file):
    full_text = ""
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            full_text += page.extract_text() or ""
    return full_text

# Function to create and save the FAISS vector database
def create_and_save_vector_db(texts, filenames, folder_path):
    # Use RecursiveCharacterTextSplitter to ensure proper chunking
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,   # Target size for each chunk
        chunk_overlap=200  # Overlap to maintain context
    )
    
    chunks = []
    metadatas = []

    for text, filename in zip(texts, filenames):
        # Split text into chunks
        split_chunks = text_splitter.split_text(text)
        chunks.extend(split_chunks)
        # Add metadata for each chunk with the source as the filename
        metadatas.extend([{"source": filename}] * len(split_chunks))

    # Create the FAISS vector store with metadata
    db = FAISS.from_texts(chunks, OpenAIEmbeddings(), metadatas=metadatas)
    
    # Save FAISS database to disk
    db.save_local(folder_path)
    print(f"FAISS vector database saved to {folder_path}")

# Directory containing your PDF files
pdf_folder = "/papers"
pdf_files = [os.path.join(pdf_folder, file) for file in os.listdir(pdf_folder) if file.endswith(".pdf")] 

# Extract text from all PDFs and keep track of filenames
all_texts = []
all_filenames = []

for pdf_file in pdf_files:
    try:
        print(f"Processing: {pdf_file}")
        full_text = pdftotext(pdf_file)
        all_texts.append(full_text)
        all_filenames.append(os.path.basename(pdf_file))
    except Exception as e:
        print(f"Error processing {pdf_file}: {e}")

# Create and save the FAISS vector database
db_folder = "/Project/faiss_db"
create_and_save_vector_db(all_texts, all_filenames, db_folder)
```
